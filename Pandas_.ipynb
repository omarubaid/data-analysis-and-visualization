# Identify and Resolve Missing data
-----

## 1. Identifying missing data

```python
filename = '/home/omarubaid/Downloads/Ex_Files_Python_for_Data_Vis/Exercise Files/Pandas/data/table_i702t60_missing_data.xlsx'
df = pd.read_excel(filename)
```

```python
# Check if there is missing data
df.info()
```

```python
# See if missing data is top 5
df['interest_paid'].isna().head()
```

```python
# Filter out the missing value and locate

interest_missing = df['interest_paid'].isna()
df.loc[interest_missing, :]
```

```python
# not operator (~) to show all rows except missing data containing row 
df.loc[~interest_missing, :]
```

```python
# removing the row might not be a good solution
df[30:40].dropna (how = 'any')
```

## 2. Filling in the missing value
Variety of ways, some are applied here;

#### â€¢ with the actual value (if known or possible to know )

```python
interest_missing = df['interest_paid'].isna()
df.loc[interest_missing, 'interest_paid'] = 93.24
```

####  â€¢ with 0

```python
df['interest_paid'][30:40].fillna(0)
```

#### â€¢ with back-fill method

```python
df['interest_paid'][30:40].fillna(method = 'bfill')
```

#### â€¢ with forward-fill method

```python
df['interest_paid'][30:40].fillna(method = 'ffill')
```

#### â€¢ with linear interpolation

```python
df['interest_paid'][30:40].interpolate(method = 'linear')
```
