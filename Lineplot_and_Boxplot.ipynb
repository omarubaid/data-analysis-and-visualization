### Convert Pandas DataFrame into Numpy Array

```python
month_number = df.loc[:, 'month'].values
interest_paid = df.loc[:, 'interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

```python
type(month_number), type(interest_paid), type(principal_paid)
```

### simple plot method  


```python
plt.plot(month_number, interest_paid)
plt.plot(month_number, principal_paid)
```

## Marker types, marker size, and colors

Below are three same graphs using color parameter as string, hex-string and tuple.

```python
plt.style.use('classic')

plt.plot(month_number, interest_paid, c='k', marker='.', markersize= 10)
plt.plot(month_number, principal_paid, c='b', marker='.', markersize= 10)
```

```python
plt.style.use('classic')

plt.plot(month_number, interest_paid, c='#000000', marker='.', markersize= 10)
plt.plot(month_number, principal_paid, c='#0000FF', marker='.', markersize= 10)
```

```python
plt.style.use('classic')

plt.plot(month_number, interest_paid, c=(0, 0, 0), marker='.', markersize= 10)
plt.plot(month_number, principal_paid, c=(0, 0, 1), marker='.', markersize= 10)
```

---
## Setting title, labels, limits, ticks, fontsize and grid

```python
plt.plot(month_number, interest_paid, c= 'k', marker= '.')
plt.xlabel('Month', fontsize = 15)
plt.xlim(0, 70)
plt.xticks(fontsize= 12)

plt.plot(month_number, principal_paid, c= 'b', marker= '.')
plt.ylabel('Dollars', fontsize = 15)
plt.ylim(0, 700)
plt.yticks(fontsize= 14)

plt.title('Interest and Principal Paid Each Month', fontsize = 18)
plt.grid(c= 'g',     #for grid only horizontally or vertically, you can set axis= 'x' or axis= 'y', respectively.
         alpha=.5,
         linestyle= '-')
```

---
## Plot Legends
---

### • within the plot

```python
plt.plot(month_number, interest_paid, c='k', label = 'Interest')
plt.plot(month_number, principal_paid, c='b', label = 'Principal')
plt.legend(loc= "center right")
```

### • outside the plot

```python
plt.plot(month_number, interest_paid, c='k', label = 'Interest')
plt.plot(month_number, principal_paid, c='b', label = 'Principal')
plt.legend(loc= (1.02, 0.5))
```

 ----
 ## Saving plot as image and adjusting its layout
 ----

```python
plt.figure(figsize=(10, 5))

plt.plot(month_number, interest_paid, c= 'k', label = 'Interest')
plt.xlabel('Month', fontsize = 15)
plt.xlim(0, 70)
plt.xticks(fontsize= 12)

plt.plot(month_number, principal_paid, c= 'b', label = 'Principal')
plt.ylabel('Dollars', fontsize = 15)
plt.ylim(0, 700)
plt.yticks(fontsize= 14)

plt.title('Interest and Principal Paid Each Month', fontsize = 18)
plt.legend(loc=(1.02, 0.5))
plt.grid()

plt.tight_layout()   #this line fixed the cut out area of legend 
plt.savefig('Desktop/graphplot.png')
```

# Boxplot
----

### 1. Using Matplotlib

```python
cancer_df = pd.read_csv('/home/omarubaid/Downloads/Ex_Files_Python_for_Data_Vis/Exercise Files/Matplotlib/data/wisconsinBreastCancer.csv')
```

```python
cancer_df.head()
```

```python
malignant = cancer_df.loc[cancer_df['diagnosis'] == 'M', 'area_mean'].values
benign = cancer_df.loc[cancer_df['diagnosis' ]== 'B', 'area_mean'].values
```

```python
plt.boxplot([malignant, benign], labels = ['M', 'B'])
```

### 2. Using Pandas as Matplotlib wrapper

```python
cancer_df.boxplot(column = 'area_mean', by = 'diagnosis');
plt.title('');
plt.suptitle('')
```

### 3. Using Seaborn

```python
sns.boxplot(x = 'diagnosis', y = 'area_mean', data = cancer_df)
```
